```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/machine_learning_topics/RemovingOutliers/heights.csv')
df.head()
```

```python
# Detect outliers using percentile
max_threshold = df['height'].quantile(0.95)
max_threshold
```

```python
df[df['height']>max_threshold]
```

```python
min_threshold = df['height'].quantile(0.05)
min_threshold
```

```python
# Removing Outliers
df[(df['height']<max_threshold) & (df['height']>min_threshold)]
```

```python
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/machine_learning_topics/DecisionTree/salaries.csv')
df.head()
```

```python
inputs = df.drop('salary_more_then_100k',axis='columns')
```

```python
target = df['salary_more_then_100k']
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
# because python can't read strings, must recode them
le_company = LabelEncoder()
le_job = LabelEncoder()
le_degree = LabelEncoder()
```

```python
inputs['cmpany_n'] = le_company.fit_transform(inputs['company'])
inputs['job_n'] = le_job.fit_transform(inputs['job'])
inputs['degree_n'] = le_degree.fit_transform(inputs['degree'])
```

```python
inputs
```

```python
inputs_n = inputs.drop(['company', 'job', 'degree'],axis='columns')
```

```python
inputs_n
```

```python
target
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(inputs_n, target)
```

```python
model.score(inputs_n,target)
```

```python
# Is salary of Google, Computer Engineer, Bachelors degree > 100 k ?
model.predict([[2,1,0]])
```

```python
# Is salary of Google, Computer Engineer, Masters degree > 100 k ?
model.predict([[2,1,1]])
```
